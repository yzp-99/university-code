```python
import pandas as pd
from pandas.tseries import offsets
from matplotlib import pyplot as plt

# 第一题
f = open('wu.csv')
data = pd.read_csv(f)
# print(data.head())


# 第二题
'''
    不知道函数里的参数之前的方法
    index = pd.to_datetime([str(i) for i in data['交易日期'].values.tolist()])
    index_1 = [str(i)[:10] for i in index.values]
    index_2 = index_1[::-1] # 做了个倒序（把时间从小到大排序）
'''
index_2 = pd.to_datetime(data['交易日期'].values.tolist(), format='%Y%m%d')
data.index = index_2
print(data.head())


# 第三题
plt.rcParams['font.sans-serif'] = ['SimHei'] # 显示中文
plt.rcParams['font.size'] = 7.5 # 条件显示图像的行列字体大小
plt.rcParams['axes.unicode_minus'] = False # 显示负号

plt.title('开盘价')
data['开盘价'].plot()
plt.show()


# 第四题:
data_copy = data
# print(data_copy.head())
t = data_copy['2017':'2014']['开盘价']
# print(t)

plt.title('2014-2017年收盘价均值')
plt.plot(t.resample('W-MON').mean())
plt.show()
```
