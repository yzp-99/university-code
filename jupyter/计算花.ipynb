```python
import numpy as np #导包
```

```python
data = np.loadtxt("iris_sepal_length.csv",delimiter=",") #读数据
print(data)
```

```python
data = np.sort(data) #排序
print(data)
```

```python
data2 = np.unique(data) #去重
print(data2)
```

```python
print(np.sum(data)) #求总长度
print(np.cumsum(data)) #求累加的和
print(np.mean(data)) #求均值
print(np.var(data)) #求方差
print(np.std(data)) #求标准差
print(np.max(data)) #求标准差
print(np.min(data)) #求标准差
```
