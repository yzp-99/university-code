```python
import pandas as pd
import numpy as np
```

```python
file = open("sport.csv")
df = pd.read_csv(file)
df
```

```python
lqdf = df.groupby('项目')
for i in lqdf:
    print(i)
```

```python
LQDT = dict(key for key in lqdf)['篮球']
LQDT.mean()
```

```python
SEXLQDT = LQDT.groupby(by='性别')
for i in SEXLQDT:
    print(i)
```

```python
maleDF = dict(key for key in SEXLQDT)['男']
maleDF
```

```python
def getMaxMinDelta(arr):
    return arr.max() - arr.min()
```

```python
maleDF.agg({'年龄（岁）':getMaxMinDelta,
            '身高(cm)':getMaxMinDelta,
            '体重(kg)':getMaxMinDelta
           })
```

```python
def getBMI(df):
    #print(df)
    height = df['身高(cm)']/100
    weight = df['体重(kg)']
    bmi = weight / height **2
    return bmi
```

```python
LQDT.apply(getBMI,axis=1)
```
